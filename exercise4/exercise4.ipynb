#### 5.4 Compare K means and GMM; what is their principal difference? How their outputs differ here and what might be the reason for it?

From the results it's clear that the GMM performs better than Kmeans. GMM works by assigning the probabilities to datapoints. Kmeans cluster assignment is a hard assignment where we are certain that each data point belongs to a particular cluster. But in GMM, it calculates how likely the datapoint will belong to each cluster. Therefore we can clearly see how correct the prediction of the cluster of a datapoint (highest probability).
